```python
from tkinter import *
        
        
cnt=0
KnightCnt=0
KnightPrice=10
ArcherCnt=0
ArcherPrice=100
Total=100
DPS=0

win=Tk()#ウィンドウ作成
win.title('title')
win.geometry('400x300')

staticK=Label(text=str(KnightCnt),foreground='red',background='yellow')#ウインドウにテキスト表示
#foreground 文字色
#background 背景色
staticK.place(x=120,y=50)#表示位置変更

staticA=Label(text=str(ArcherCnt),foreground='red',background='yellow')#ウインドウにテキスト表示
staticA.place(x=120,y=100)#表示位置変更

staticT=Label(text=str(Total),foreground='red',background='white')#ウインドウにテキスト表示
staticT.place(x=10,y=10)#表示位置変更
staticT.pack()#真ん中に表示

staticD=Label(text=str(DPS),foreground='red',background='white')#ウインドウにテキスト表示
staticD.place(x=10,y=10)#表示位置変更
staticD.pack()#真ん中に表示

def function(self):
    global KnightCnt
    global KnightPrice
    global Total
    if(Total>KnightPrice):
        Total -=KnightPrice
        KnightCnt+=1
        KnightPrice=int(KnightPrice*1.1)
    staticK["text"]=str(KnightCnt)
    ButtonK["text"]='Knight \\'+str(KnightPrice)
    
def ArcFunc(self):
    global ArcherCnt
    global Total
    global ArcherPrice
    if(Total>ArcherPrice):
        Total -=ArcherPrice
        ArcherCnt+=1
        ArcherPrice=int(ArcherPrice*1.1)
    staticA["text"]=str(ArcherCnt)
    ButtonA["text"]='Archer \\'+str(ArcherPrice)
    

ButtonK=Button(text='Knight \\'+str(KnightPrice))#ボタン作成
ButtonK.bind("<Button-1>",function)
ButtonK.place(x=150,y=50)
             
ButtonA=Button(text='Archer \\'+str(ArcherPrice))#ボタン作成
ButtonA.bind("<Button-1>",ArcFunc)
ButtonA.place(x=150,y=100)

ValK=BooleanVar()
ValA=BooleanVar()
ValK.set(True)
ValA.set(True)

CheckBoxK=Checkbutton(text='Knight',variable=ValK)
CheckBoxK.place(x=50,y=50)
CheckBoxA=Checkbutton(text='Archer',variable=ValA)
CheckBoxA.place(x=50,y=100)

def game_loop():
    global Total
    global DPS
    DPS=1+KnightCnt+ArcherCnt*10
    Total += DPS
    staticT["text"]='Total ' + str(Total)
    staticD["text"]='DPS ' +str(DPS)
    if Total<10000:
        win.after(300,game_loop)#300ms後に再度実行(再帰)
    if Total>10000:
        print('Clear!!')
    
    
game_loop()
win.mainloop()#ウインドウ作成とセット
```

```python
from tkinter import *

#キャラクター
class Chara:
    
    #初期設定
    def __init__(self,name,count,price,xplace,yplace):
        self.name=name
        self.count=count
        self.price=int(price)
        
        #ラベル情報
        self.s=Label(text=str(self.count),foreground='red',background='yellow')
        self.s.place(x=xplace,y=yplace)
        
        #ボタン情報
        self.b=Button(text=self.name+ '\\'+str(self.price))#ボタン作成
        self.b.place(x=xplace+50,y=yplace)
        self.b.bind("<Button-1>",self.buying)
        
        
        #ラベル情報
        self.s=Label(text=str(self.count),foreground='red',background='yellow')
        self.s.place(x=xplace,y=yplace)
        
    #購入時の処理
    def buying(self,total):
        global Total
        if Total.count>self.price:
            Total.count-=self.price
            self.count+=1
            self.price=int(self.price*1.1)
        self.s["text"]=str(self.count)
        self.b["text"]=self.name + '\\' +str(self.price)
    
class Info:
    def __init__(self,name,count,xplace,yplace):
        self.name=name
        self.count=count
        
        #ラベル情報
        self.s=Label(text=str(self.count),foreground='red',background='yellow')
        self.s.place(x=xplace,y=yplace)
        
#ウィンドウ作成
win=Tk()
win.title('title')
win.geometry('400x300')
    
#キャラクター設定
Knight=Chara('Knight',0,10,100,100)
Archer=Chara('Archer',0,100,100,150)
Ninja=Chara('Ninja',0,1000,100,200)

Total=Info('Total',100,150,10)
DPS=Info('DPS',1,150,30)

#ゲームループ
def game_loop():
    DPS.count=1+Knight.count+Archer.count*10
    Total.count += DPS.count
    Total.s['text']=Total.name + ' \\' +str(Total.count)
    DPS.s['text']=DPS.name + ' \\' +str(DPS.count)

    if Total.count<10000:
        win.after(300,game_loop)#300ms後に再度実行(再帰)
    if Total.count>10000:
        print('Clear!!')
    
game_loop()
win.mainloop()#ウインドウ作成とセット
```
